## Statistical_Analysis

The following code will create a table from the species data which will show the amount of each species category in each conservation status category. From this we will pick two species categories and two conservation status category and perform a chi-square contigency test to determine if there is any statisical significance.

```python
from scipy.stats import chi2_contingency
cat_constat = species.groupby(["category","conservation_status"]).scientific_name.nunique().reset_index().pivot(columns='conservation_status',
                                      index='category',
                                      values='scientific_name').reset_index()
cat_constat.columns = ['category', 'Endangered', 'In Recovery','Species of Concern','Threatened']
cat_constat=cat_constat.fillna(0)
print(cat_constat)
contingency = [[68, 0],
               [4, 2]]
chi2_contingency(contingency)
```

### Chi-Square_Contingency_Test

||Species of Concern|Threatened|
|-|-|-|
|Bird|68|0|
|Amphibian|4|2|


p-value: 0.0004422475247430433

Given that the p-value is much smaller than the common significance level of 0.05, the null hypothesis can be rejected. This means there is a significant association between 'Category' (Bird vs. Amphibian) and 'Conservation Status' (Species of Concern vs. Threatened).

In other words, the distribution of species between Species of Concern and Threatened categories is significantly different for Birds and Amphibians.

### Conclusion

This work analyzed data from the National Parks Service on endangered species in different parks to investigate patterns in the types of species that become endangered. The analysis focused on the conservation statuses of various species. The study found a significant association between the category of a species (bird vs. amphibian) and its conservation status (species of concern vs. threatened). The distribution of species between these conservation statuses was significantly different for birds and amphibians. The p-value obtained from the chi-square contingency test was much smaller than the significance level of 0.05, leading to the rejection of the null hypothesis. This finding suggests that birds are more likely to be listed as a species of concern, while amphibians are more likely to be listed as threatened.
